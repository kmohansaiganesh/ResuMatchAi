## 1.IMPORTS

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
```

## 2. FUNCTIONS

```python
def extract_text_from_url_selenium(url):
    global driver
    cleaned_text = "Extracted using Selenium"

    # Open the channel URL
    driver.get(url)

    # Wait until the body of the page has loaded, or timeout after 10 seconds
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.TAG_NAME, "body"))   
    )

    #PENDING --- try to reduce/ remove the sleep
    time.sleep(5)

    # # Extract text from the entire page
    text = driver.find_element(By.TAG_NAME, 'body').text

    # Handle encoding gracefully
    try:
        cleaned_text = text.encode('utf-8').decode('utf-8', errors='ignore')
    except UnicodeEncodeError as e:
        print("Encoding error encountered:", e)

    return cleaned_text.strip()
```

## 3. USAGE

```python
driver = webdriver.Chrome()
```

```python
url = "https://jobs.careers.microsoft.com/global/en/job/1875035/%C2%A0Software-Development-Engineer-II"
text_content = extract_text_from_url_selenium(url)
print(text_content)
```
